```python
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

# --- 1. CORE CONSTANTS ---
G, c, h = 6.67430e-11, 299792458, 6.62607e-34
RHO_CRITICAL = 6e-27  # kg/m^3 (Dark Energy Floor)
date_stamp = "March 30, 2026"

# Planck Scale (Quantum Gravity Crossover)
r_planck, m_planck = 1.616e-35, 2.176e-8

plt.figure(figsize=(24, 18), facecolor='#121212')
ax = plt.gca()
ax.set_facecolor('#121212')
r_grid = np.logspace(-45, 35, 1000)
plt.grid(True, which="both", ls="-", color='#333333', alpha=0.3)

# --- 2. DARK UNIVERSE LAYERS ---
# Dark Matter (NFW Halo Profile)
dm_r = np.logspace(20, 24, 100)
dm_m = 1e42 * (dm_r/1e21)**0.8
plt.loglog(dm_r, dm_m, color='#ffff00', ls='--', lw=4, label='DARK MATTER (Galaxy Halos)', alpha=0.9, zorder=30)

# Dark Energy (Uniform Floor)
de_m = RHO_CRITICAL * (4/3 * np.pi * r_grid**3)
plt.loglog(r_grid, de_m, color='#bf00ff', lw=4, ls='-', label='DARK ENERGY (Uniform)', alpha=0.7, zorder=20)
plt.fill_between(r_grid, 1e-48, de_m, color='#bf00ff', alpha=0.03)

# --- 3. QUANTUM GRAVITY (QG) ZONE ---
plt.axvspan(1e-40, 1e-30, color='lime', alpha=0.08, label='QG REGIME (Unification)')
plt.scatter(r_planck, m_planck, c='lime', marker='*', s=1500, ec='white', lw=3, zorder=120, label='PLANCK CROSSOVER (QG)')

# --- 4. PHYSICS BOUNDARIES ---
plt.loglog(r_grid, (c**2 * r_grid)/(2*G), color='#ffffff', lw=4, ls='--', label='SCHWARZSCHILD (Gravity)', zorder=40)
plt.loglog(r_grid, h/(r_grid*c), color='#ff00ff', lw=4, ls=':', label='COMPTON (Quantum)', zorder=40)

# --- 5. LATTICELOG OBJECTS & LEVERS ---
points = {
    "Atoms": (1e-10, 1e-26, "#00ffff"), "Planets": (1e7, 1e25, "#00ffff"),
    "Stars": (1e9, 1e30, "#00ffff"), "Galaxies": (1e21, 1e42, "#00ffff"),
    "Clusters": (1e23, 1e45, "#00ffff"), "MICRO (11)": (r_planck, m_planck, "#ffff00"),
    "FLEA (30)": (1e-3, 1e-5, "#ffff00"), "LUNGS": (1, 70, "#ffff00"),
    "MACRO (19)": (4.4e26, 1e54, "#ffff00")
}

for name, (r, m, color) in points.items():
    plt.scatter(r, m, c=color, s=500, edgecolors='white', zorder=105)
    y_off = 10**3.8 if "MACRO" in name else 10**1.8
    plt.text(r, m * y_off, name, color=color, fontsize=16, fontweight='bold', ha='center',
             bbox=dict(boxstyle="round,pad=0.2", facecolor='#121212', alpha=0.8, edgecolor='none'))

# Universe Gold Star
plt.scatter(4.4e26, 1e54, c='#ffd700', marker='*', s=5000, ec='white', lw=2, zorder=110)

# --- 6. PHYSICS ANNOTATIONS (Textbook Labels) ---
plt.annotate('DM Halo Jump\\n1e42 * (r/1e21)^0.8', xy=(1e21, 1e42), xytext=(1e15, 1e55),
             arrowprops=dict(arrowstyle='->', color='#ffff00', lw=2),
             fontsize=16, color='#ffff00', fontweight='bold')

plt.annotate('ΛCDM DE Floor\\n6e-27 kg/m³', xy=(1e27, 1e60), xytext=(1e18, 1e85),
             arrowprops=dict(arrowstyle='->', color='#bf00ff', lw=2),
             fontsize=16, color='#bf00ff', fontweight='bold')

# Branding & Axis
plt.title(f"LatticeLogOS: Unified Planck-to-Hubble Scale | {date_stamp}", fontsize=34, color='#ffd700', pad=30)
plt.xlabel("Scale (Meters, log)", fontsize=22, color='white')
plt.ylabel("Mass (Kilograms, log)", fontsize=22, color='white')
plt.tick_params(colors='white', labelsize=18)
plt.xlim(1e-45, 1e35); plt.ylim(1e-48, 1e110)
plt.legend(loc='upper left', fontsize=16, frameon=True, facecolor='#121212', edgecolor='white', labelcolor='white')

plt.tight_layout()
plt.savefig('latticelog_final_unification.png', dpi=300, facecolor='#121212')
print("✅ Chart Generated Successfully!")
files.download('latticelog_final_unification.png')
```
