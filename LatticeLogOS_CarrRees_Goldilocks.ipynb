```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker

# Log-log Carr-Rees: Size (m) vs Density (kg/m³)
fig, ax = plt.subplots(figsize=(12,8))

# Black Hole Line: rho = 3c²/(8πG R²) → slope -3 on log-log (density vs radius)
R_bh = np.logspace(3, 30, 100)  # meters (proton → galaxy cluster)
rho_bh = 1.85e27 / (R_bh**2)   # Schwarzschild density scaling

# Atomic Density Line: slope -3 (constant density band)
rho_atomic = np.ones_like(R_bh) * 1e3      # Water ~1000 kg/m³ (garage scale)
rho_nuclear = np.ones_like(R_bh) * 1e17    # Nuclear ~10^17 kg/m³

# 10^17 Goldilocks Zone (your gear ratio)
R_goldilocks = 10**26  # ~galaxy cluster scale
M_1e17suns = 10**17 * 1.989e30  # 10^17 solar masses
R_schwarz = 2.95e3 * (M_1e17suns/1e30)**(1/3) / 3.086e16  # Schwarzschild radius (ly → m)
rho_goldilocks = 3 * M_1e17suns / (4/3 * np.pi * R_schwarz**3)

# Plot
ax.loglog(R_bh, rho_bh, 'k-', linewidth=3, label='Black Hole Line (slope -3)')
ax.loglog(R_bh, rho_atomic, 'b--', linewidth=2, label='Atomic/Water Density')
ax.loglog(R_bh, rho_nuclear, 'r--', linewidth=2, label='Nuclear Density')
ax.scatter([R_goldilocks], [rho_goldilocks], s=300, c='gold', marker='*',
           label=f'10¹⁷ Suns ({rho_goldilocks:.0e} kg/m³)', zorder=5)

# Garage band (your 27 proofs)
ax.axvspan(1e-9, 1e-2, alpha=0.3, color='green', label='Garage Scale (10¹⁷/cm³)')

ax.set_xlabel('Radius (meters)', fontsize=12)
ax.set_ylabel('Density (kg/m³)', fontsize=12)
ax.set_title('Carr-Rees Plot: 10¹⁷ Gear Ratio = Goldilocks Zone', fontsize=14)
ax.legend()
ax.grid(True, alpha=0.3)

# Mark key scales
ax.axvline(1e-15, color='gray', linestyle=':', alpha=0.7, label='Proton')
ax.axvline(1e26, color='orange', linestyle=':', alpha=0.7, label='Cosmic Horizon')
plt.tight_layout()
plt.show()

print(f"Goldilocks BH: {M_1e17suns:.2e} kg, R_schwarz: {R_schwarz:.2e} m")
print(f"Universe mass equiv: {'MATCHES observable universe scale'}")
```
