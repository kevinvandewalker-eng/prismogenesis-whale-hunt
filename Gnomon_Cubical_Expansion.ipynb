```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from google.colab import files

def draw_cube(ax, offset, size, color, alpha):
    r = [0, size]
    for s, e in [(i, j) for i in r for j in r]:
        ax.plot([s+offset[0], s+offset[0]], [e+offset[1], e+offset[1]], [0+offset[2], size+offset[2]], color=color, alpha=alpha, lw=2)
        ax.plot([s+offset[0], s+offset[0]], [0+offset[1], size+offset[1]], [e+offset[2], e+offset[2]], color=color, alpha=alpha, lw=2)
        ax.plot([0+offset[0], size+offset[0]], [s+offset[1], s+offset[1]], [e+offset[2], e+offset[2]], color=color, alpha=alpha, lw=2)

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('#121212')
fig.patch.set_facecolor('#121212')

# Draw 4x4x4 (Outer Expansion) first so it sits in back
draw_cube(ax, [0, 0, 0], 4, '#bf00ff', 0.3)

# Draw 3x3x3 (Inner Universe) centered at 0.5 offset
draw_cube(ax, [0.5, 0.5, 0.5], 3, '#00ffff', 0.8)

# Add the "Spheres" (Lattice Nodes) at every integer intersection
for x in range(5):
    for y in range(5):
        for z in range(5):
            ax.scatter(x, y, z, color='gold', s=50, alpha=0.8, edgecolors='white', lw=0.5)

ax.set_title("LatticeLogOS: Gnomon Cubical Expansion (3³ → 4³)", color='gold', fontsize=22, pad=20)
ax.axis('off')
ax.view_init(elev=20, azim=45)

# Save and Trigger Download
plt.savefig('gnomon_expansion.png', dpi=300, bbox_inches='tight', facecolor='#121212')
print("✅ Gnomon Expansion Diagram Generated")
files.download('gnomon_expansion.png')
```
