```python
import matplotlib.pyplot as plt
import numpy as np

def lattice_10_20_suite():
    fig, axes = plt.subplots(2, 3, figsize=(24, 12), facecolor='white')

    # A. WHALE TISSUE (10²⁰ cells) - FIXED BOUNDS
    rows, cols = 25, 25
    for r in range(rows):
        for c in range(cols):
            x = (c-cols/2)*1.2 + np.random.uniform(-0.08,0.08)
            y = (r-rows/2)*np.sqrt(3) + np.random.uniform(-0.08,0.08)
            if c % 2: y += np.sqrt(3)/2
            axes[0,0].scatter(x, y, s=80, marker='H', c='darkblue',
                            edgecolors='navy', alpha=0.8)
    axes[0,0].set_xlim(-18, 18); axes[0,0].set_ylim(-22, 22)
    axes[0,0].set_title("🐋 WHALE\n10²⁰ Cells", fontweight='bold', fontsize=14)
    axes[0,0].axis('off')

    # B1. CATFISH TASTE BUDS
    for r in range(20):
        for c in range(20):
            x = (c-10)*1.4 + np.sin(r*0.3)*0.3
            y = (r-10)*1.0 + np.cos(c*0.4)*0.2
            axes[0,1].scatter(x, y, s=100, c='orange', marker='o')
    axes[0,1].set_xlim(-15, 15); axes[0,1].set_ylim(-12, 12)
    axes[0,1].set_title("🐟 CATFISH\n10²⁰ Taste Buds", fontweight='bold', fontsize=14)
    axes[0,1].axis('off')

    # B2. DIATOM SHELLS
    for r in range(22):
        for c in range(22):
            x = (c-11)*1.1
            y = (r-11)*np.sqrt(3)*0.9
            if c % 2: y += np.sqrt(3)*0.45
            axes[0,2].scatter(x, y, s=90, marker='H', c='lightblue')
    axes[0,2].set_xlim(-13, 13); axes[0,2].set_ylim(-16, 16)
    axes[0,2].set_title("🌊 DIATOMS\n10²⁰ O₂ Engine", fontweight='bold', fontsize=14)
    axes[0,2].axis('off')

    # C1. SOIL MICROBES
    for i in range(400):
        theta = i * 0.02
        r = np.sqrt(i) * 0.4
        x, y = r*np.cos(theta), r*np.sin(theta)
        axes[1,0].scatter(x, y, s=70, c='brown', alpha=0.7)
    axes[1,0].set_title("🌱 SOIL\n10²⁰ Microbes", fontweight='bold', fontsize=14)
    axes[1,0].axis('off')

    # C2. CHLOROPLASTS
    for r in range(18):
        for c in range(18):
            x = (c-9)*1.3
            y = (r-9)*1.1
            if c % 2: x += 0.65
            axes[1,1].scatter(x, y, s=110, c='green', marker='s')
    axes[1,1].set_xlim(-12, 12); axes[1,1].set_ylim(-11, 11)
    axes[1,1].set_title("☀️ CHLOROPLASTS\n10²⁰ Solar", fontweight='bold', fontsize=14)
    axes[1,1].axis('off')

    # C3. WHALE SYNAPSES
    for r in range(23):
        for c in range(23):
            x = (c-11.5)*1.0 + np.random.uniform(-0.15,0.15)
            y = (r-11.5)*np.sqrt(3)*0.85 + np.random.uniform(-0.15,0.15)
            if c % 2: y += np.sqrt(3)*0.425
            axes[1,2].scatter(x, y, s=60, marker='H', c='purple')
    axes[1,2].set_xlim(-13, 13); axes[1,2].set_ylim(-16, 16)
    axes[1,2].set_title("🧠 SYNAPSES\n10²⁰ Neural Grid", fontweight='bold', fontsize=14)
    axes[1,2].axis('off')

    plt.suptitle("LatticeLogOS: 10²⁰ CRITICAL PACKING LAYER\nEvolution's Lattice Limit",
                 fontsize=20, fontweight='bold', y=0.95)
    plt.tight_layout()
    plt.savefig('10_20_Critical_Layer.png', dpi=400, bbox_inches='tight')
    plt.show()

lattice_10_20_suite()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def critical_packing_layer():
    # Adjusted figsize for better vertical breathing room
    fig, axes = plt.subplots(2, 3, figsize=(18, 10), facecolor='white')

    # --- HEADER FIX ---
    # Using suptitle with a high y-value and manual line breaks
    fig.suptitle("LatticeLogOS: $10^{20}$ CRITICAL PACKING LAYER\nEvolution's Lattice Limit",
                 fontsize=18, fontweight='bold', y=0.96)

    # 1. WHALE (Top Left)
    x, y = np.meshgrid(np.linspace(0, 10, 20), np.linspace(0, 10, 20))
    axes[0,0].scatter(x, y, c='darkblue', s=30, alpha=0.7)
    axes[0,0].set_title("🐋 WHALE\n$10^{20}$ Cells", fontweight='bold', pad=15)
    axes[0,0].axis('off')

    # 2. CATFISH (Top Middle) - Fixed title overlap
    x_cat = np.linspace(0, 10, 20)
    y_cat = np.linspace(0, 10, 40)
    X, Y = np.meshgrid(x_cat, y_cat)
    X_wave = X + 0.2 * np.sin(Y) # Wavy pattern
    axes[0,1].scatter(X_wave, Y, c='orange', s=25)
    axes[0,1].set_title("🐟 CATFISH\n$10^{20}$ Taste Buds", fontweight='bold', pad=15)
    axes[0,1].axis('off')

    # 3. DIATOMS (Top Right)
    x_dia = np.linspace(0, 10, 25)
    y_dia = np.linspace(0, 10, 25)
    X, Y = np.meshgrid(x_dia, y_dia)
    axes[0,2].scatter(X, Y, c='lightblue', s=40, alpha=0.6)
    axes[0,2].set_title("🔬 DIATOMS\n$10^{20}$ $O_2$ Engine", fontweight='bold', pad=15)
    axes[0,2].axis('off')

    # 4. SOIL (Bottom Left)
    theta = np.linspace(0, 10 * np.pi, 500)
    r = np.exp(0.1 * theta)
    axes[1,0].plot(r * np.cos(theta), r * np.sin(theta), c='firebrick', lw=4, alpha=0.8)
    axes[1,0].set_title("🌱 SOIL\n$10^{20}$ Microbes", fontweight='bold', pad=15)
    axes[1,0].axis('off')

    # 5. CHLOROPLASTS (Bottom Middle)
    for i in range(10):
        x_block = np.full(20, i)
        y_block = np.linspace(0, 10, 20)
        axes[1,1].scatter(x_block, y_block, marker='s', c='green', s=40)
    axes[1,1].set_title("🍃 CHLOROPLASTS\n$10^{20}$ Solar", fontweight='bold', pad=15)
    axes[1,1].axis('off')

    # 6. SYNAPSES (Bottom Right)
    x_syn = np.random.normal(5, 2, 800)
    y_syn = np.random.normal(5, 2, 800)
    axes[1,2].scatter(x_syn, y_syn, c='purple', s=15, alpha=0.6)
    axes[1,2].set_title("🧠 SYNAPSES\n$10^{20}$ Neural Grid", fontweight='bold', pad=15)
    axes[1,2].axis('off')

    # TIGHT_LAYOUT FIX: rect=[left, bottom, right, top]
    # This keeps the subplots from moving into the header space
    plt.tight_layout(rect=[0, 0.03, 1, 0.92], pad=3.0)

    # Save the cleaned version
    plt.savefig('Critical_Packing_10_20.png', dpi=300, bbox_inches='tight')
    plt.show()

critical_packing_layer()
```
