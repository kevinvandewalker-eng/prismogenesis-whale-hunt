```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(12, 8))

# Data
scales = ['Planck Chaos', 'LUCA Genome\n(2.75 Mb)', 'LUCA Proteins\n(H(19)=2657)',
          'CRISPR Limit', 'Whale Cells', 'BH Floor']
values = [1e-35, 2.75e6, 2657, 1e17, 1e20, 1e17]
log_values = np.log10(values)

# Colors
colors = ['black', 'purple', 'gold', 'cyan', 'blue', 'red']

bars = ax.bar(scales, log_values, color=colors, edgecolor='white', linewidth=2)
ax.set_ylabel('Log$_{10}$ Scale', fontsize=12)
ax.set_title('LatticeLogOS: LUCA → 10$^{20}$ → 10$^{17}$ Convergence', fontsize=16, fontweight='bold')
ax.axhline(y=np.log10(1e17), color='red', linestyle='--', linewidth=3, label='10$^{17}$ Lattice Floor')

# Bar value labels - Logic to prevent overlap with axis
for bar, val in zip(bars, log_values):
    height = bar.get_height()
    va_dir = 'bottom' if val >= 0 else 'top'
    offset = 0.5 if val >= 0 else -0.5
    ax.text(bar.get_x() + bar.get_width()/2., height + offset,
            f'{val:.1f}', ha='center', va=va_dir, fontweight='bold')

# LUCA H(19) highlight - MOVED higher with an arrow
ax.annotate('H(19) = 2657\nLUCA Proteins\nExact Match!',
            xy=(2, 3.42),      # Pointing to the top of the bar
            xytext=(2, 12),     # Moved text much higher (Y=12)
            bbox=dict(boxstyle='round', facecolor='gold', alpha=0.8),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
            fontsize=12, ha='center', fontweight='bold')

plt.legend()
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
