```python
import numpy as np
import matplotlib.pyplot as plt

# N52 rotating B-field + salt response
r = np.linspace(0.01, 0.1, 100)  # salt grain distances
omega_table = 35  # Hz spin
B_n52 = 1.45  # Tesla static field

# Frame-dragging torque ∝ ω × B
drag_force = B_n52 * omega_table * r**2
hex_distort = np.sin(2*np.pi*omega_table*t + drag_force)

plt.plot(r, hex_distort)
plt.title("Kerr Twist vs Salt Lattice")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# N52 spinning table parameters
r = np.linspace(0.01, 0.1, 100)  # salt grain distances (m)
omega_table = 35  # table spin (Hz = rad/s * 2pi, but keep Hz for now)
B_n52 = 1.45  # N52 magnet field strength (Tesla)
t = np.linspace(0, 1, 100)  # time array (1 second)

# Static case (Schwarzschild analog)
hex_static = np.sin(2*np.pi*35*t[:, None] + r[None, :])

# Spinning case (Kerr frame-dragging analog)
drag_torque = B_n52 * omega_table * r**2 * 0.01  # torque ∝ ωB r²
hex_spinning = np.sin(2*np.pi*35*t[:, None] + r[None, :] + drag_torque[None, :])

# Plot comparison
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(hex_static, aspect='auto', cmap='RdYlBu')
plt.title('Static Magnets\n(Schwarzschild - Clean Hex)')
plt.ylabel('Time')
plt.xlabel('Salt Distance')

plt.subplot(1, 2, 2)
plt.imshow(hex_spinning, aspect='auto', cmap='RdYlBu')
plt.title('Spinning N52 Table\n(Kerr - Frame-Dragging Twist)')
plt.ylabel('Time')
plt.xlabel('Salt Distance')

plt.tight_layout()
plt.savefig('output/kerr_test.png')  # Your video thumbnail
plt.show()

print("Static: Clean hex lattice")
print("Spinning: Expected spiral distortion")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Create output directory if it doesn't exist
os.makedirs('output', exist_ok=True)

# N52 spinning table parameters
r = np.linspace(0.01, 0.1, 100)  # salt grain distances (m)
omega_table = 35  # table spin (Hz)
B_n52 = 1.45  # N52 magnet field strength (Tesla)
t = np.linspace(0, 1, 100)  # time array (1 second)

# Static case (Schwarzschild analog) - CLEAN HEX
hex_static = np.sin(2*np.pi*35*t[:, None] + r[None, :])

# Spinning case (Kerr frame-dragging analog) - TWISTED SPIRAL
drag_torque = B_n52 * omega_table * r**2 * 0.01  # torque ∝ ωB r²
hex_spinning = np.sin(2*np.pi*35*t[:, None] + r[None, :] + drag_torque[None, :])

# Plot comparison for your video
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(hex_static, aspect='auto', cmap='RdYlBu', extent=[0, 0.1, 1, 0])
plt.colorbar(label='Phase')
plt.title('Static N52 (Schwarzschild)\nClean Hex Lattice')
plt.ylabel('Time (s)')
plt.xlabel('Distance (m)')

plt.subplot(1, 2, 2)
plt.imshow(hex_spinning, aspect='auto', cmap='RdYlBu', extent=[0, 0.1, 1, 0])
plt.colorbar(label='Phase')
plt.title('Spinning Table (Kerr)\nFrame-Dragging Spiral')
plt.ylabel('Time (s)')
plt.xlabel('Distance (m)')

plt.tight_layout()
plt.savefig('output/kerr_test.png', dpi=300, bbox_inches='tight')  # Video gold
plt.show()

print("✅ Static: Clean hexagonal lattice (Schwarzschild match)")
print("❌ Spinning: Distorted spirals (Kerr prediction)")
print("🎥 Screenshot 'output/kerr_test.png' for thumbnail")
```
