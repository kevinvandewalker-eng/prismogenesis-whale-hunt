```python
import numpy as np

class LatticeLogOS:
    def __init__(self, signatures=17000):
        self.n_total = signatures
        self.backbone_gap = 10**17
        self.current_shell = 1
        self.info_density = 0
        # Saturation is reached when density hits 100% of capacity
        self.saturation_limit = 1.0

    def render_frame(self, data_flux):
        """
        Simulates the Code-Update Frequency.
        Higher flux leads to faster Dark Matter (history) accumulation.
        """
        # We scale the flux to represent a significant portion of the backbone capacity
        accumulation = (data_flux * self.n_total) / self.backbone_gap

        # Here we 'cheat' the math for the sake of the sim,
        # using a multiplier to see the shells jump.
        self.info_density += accumulation * (10**12)

        status = f"Shell {self.current_shell} | Density: {self.info_density:.4f}"

        if self.info_density >= self.saturation_limit:
            print(f"{status} -> [SATURATED]")
            self.resync_event()
        else:
            print(status)

    def resync_event(self):
        """The Möbius Resync: Resetting the frame for the next shell."""
        print("--- EXECUTING GLOBAL RESYNC (Shell Jump) ---")
        self.current_shell += 1
        self.info_density = 0 # Resetting the 'lag' or sync-error

# Running the Lifecycle Loop
universe = LatticeLogOS()
print("LatticeLogOS Lifecycle Initialized...\n")

# Simulate 10 update cycles
for i in range(10):
    universe.render_frame(data_flux=0.5)
```
