```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Schwarzschild metric geodesic equations (null, ds^2=0 for photons)
def schwarzschild_geodesic(y, tau, M, L, E, r0):
    r, phi, dr_dtau, dphi_dtau = y
    Rs = 2 * M  # Schwarzschild radius

    # Effective potential for photons (PFDT stiffening test)
    V_eff = (L**2 / r**2) * (1 - Rs / r)

    # Geodesic ODEs (radial equation)
    d2r_dtau2 = E**2 * (Rs / (2 * r**2)) + (L**2 / r**3) * (1 - Rs / r * 3/2)

    dr_dtau_new = dr_dtau
    dphi_dtau_new = L / r**2
    dphi_dtau = dphi_dtau_new
    dr_dtau = dr_dtau_new
    r_new = r + dr_dtau * 0.01
    phi_new = phi + dphi_dtau * 0.01

    return [dr_dtau, dphi_dtau, d2r_dtau2, L / r**2]

# PFDT radiation pressure test: Gaussian photon packet collapse
def kugelblitz_test(total_energy=1e30, sigma=1e3):  # J, m
    G = 6.6743e-11
    c = 3e8
    M_light = total_energy / c**2  # Mass-equivalent
    Rs = 2 * G * M_light / c**2
    print(f"Kugelblitz Rs: {Rs:.2e} m from E={total_energy:.2e} J")

    # Test photon orbit stability at 1.5 Rs (PFDT stiffening)
    r_photon = 1.5 * Rs
    L = np.sqrt(r_photon**3 * 1.5 / Rs)  # Angular momentum for unstable orbit
    E = 1.0  # Energy

    # Integrate 1000 steps
    tau = np.linspace(0, 1000, 10000)
    y0 = [r_photon, 0, 0, L / r_photon**2]
    sol = odeint(schwarzschild_geodesic, y0, tau, args=(M_light, L, E, r_photon))

    plt.plot(sol[:,1]*180/np.pi % 360, sol[:,0]/Rs)  # phi (deg) vs r/Rs
    plt.xlabel('Azimuth Angle (deg)')
    plt.ylabel('r / Rs')
    plt.title('Photon Orbit: Stable Ring vs Spiral-In')
    plt.grid(True)
    plt.savefig('kugelblitz_photon_orbit.png')
    plt.show()

    # Metric: Stable if min(r) > Rs (no spiral)
    stability = np.min(sol[:,0]) > Rs
    print(f"Orbit Stability: {'LOCKED' if stability else 'SPIRAL-IN'}")
    np.savetxt('kugelblitz_metrics.csv', sol, header='r,phi,dr,phi_dot')

# RUN TEST
if __name__ == "__main__":
    kugelblitz_test()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Proper Schwarzschild photon radial equation (effective potential method)
def photon_radial(t, u, b, Rs):
    """u = [r, dr/dt], b=impact parameter, Rs=Schwarzschild radius"""
    r, dr = u
    V_eff = (1 - Rs/r) * (1 + b**2/r**2)  # Null geodesic effective potential
    d2r = (Rs/(2*r**2)) * (3*b**2/r**2 - 1)
    return [dr, d2r]

# PFDT kugelblitz test: radiation pressure horizon
def kugelblitz_fixed(E_total=1e53, sigma=1e6):  # Solar-mass energy for real Rs
    G, c = 6.6743e-11, 3e8
    M_eq = E_total / c**2
    Rs = 2 * G * M_eq / c**2
    print(f"Kugelblitz Rs: {Rs:.1f} m (solar mass energy)")

    # Photon sphere at 1.5 Rs, critical impact parameter b=3√3 M
    r_ph = 1.5 * Rs
    b_crit = 3 * np.sqrt(3) * Rs  # Unstable circular orbit

    # Test stable orbit (PFDT stiffening) vs unstable spiral
    t_span = (0, 2000)
    u0_stable = [r_ph, 0]  # Tangential at photon sphere

    sol_stable = solve_ivp(photon_radial, t_span, u0_stable,
                          args=(b_crit, Rs), dense_output=True, rtol=1e-8)

    # Sample orbit
    t = np.linspace(0, 2000, 10000)
    z_stable = sol_stable.sol(t)

    plt.figure(figsize=(10,4))
    plt.subplot(121)
    plt.plot(t, z_stable[0]/Rs)
    plt.axhline(1.5, color='r', ls='--', label='Photon Sphere')
    plt.axhline(1.0, color='k', ls=':', label='Horizon')
    plt.xlabel('Time'); plt.ylabel('r / Rs'); plt.title('PFDT Orbit Test')
    plt.legend(); plt.grid()

    # Stability metric: does it stay > Rs?
    min_r = np.min(z_stable[0])
    stable = min_r > Rs
    print(f"Photon Orbit: {'STABLE RING (PFDT WIN)' if stable else 'SPIRAL-IN (GR)'}")
    print(f"Min r/Rs: {min_r/Rs:.3f}")

    plt.subplot(122)
    theta = b_crit / z_stable[0]  # Angular coord
    plt.plot(theta*180/np.pi % 360, z_stable[0]/Rs)
    plt.xlabel('Angle (deg)'); plt.ylabel('r / Rs')
    plt.title('Trajectory Map')
    plt.grid()

    plt.tight_layout()
    plt.savefig('kugelblitz_fixed.png', dpi=150)
    plt.show()

    # Export garage-scale metrics
    np.savetxt('garage_kugelblitz.csv',
               np.column_stack([t, z_stable[0], theta*180/np.pi]),
               header='time,r_Rs,angle_deg', delimiter=',')
    print("Metrics saved: garage_kugelblitz.csv")

# EXECUTE
if __name__ == "__main__":
    kugelblitz_fixed()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# PFDT Rigid-Rod vs GR c-lag orbit test
def orbit_dynamics(t, y, rigid=True):
    """y = [r, theta, vr, vtheta]; rigid=True tests instant stator drag"""
    r, theta, vr, vtheta = y

    GM = 1.327e20  # Solar
    c = 3e8

    # PFDT: Rigid coordinate drag (no propagation delay)
    if rigid:
        a_r = -GM/r**2 + vtheta**2/r
        a_theta = -2*vr*vtheta/r  # Coriolis stiffening
    else:
        # GR c-lag (gravitomagnetic delay approximation)
        delay = 8.3 * np.log(r/1e11) / c  # Sun-Earth light time
        a_r = -GM/r**2 * (1 + 1.5e-6*np.sin(t-delay)**2) + vtheta**2/r
        a_theta = -2*vr*vtheta/r

    return [vr, vtheta/r, a_r, a_theta]

# Test 1-year Earth orbit stability
def rigid_rod_test():
    r0 = 1.496e11  # AU
    v0 = 29780  # m/s
    y0 = [r0, 0, 0, v0]

    t_span = (0, 3.156e7)  # 1 year
    t_eval = np.linspace(0, 3.156e7, 10000)

    # PFDT rigid vs GR delayed
    sol_rigid = solve_ivp(orbit_dynamics, t_span, y0, args=(True,),
                         t_eval=t_eval, rtol=1e-10)
    sol_gr = solve_ivp(orbit_dynamics, t_span, y0, args=(False,),
                      t_eval=t_eval, rtol=1e-10)

    plt.figure(figsize=(12,5))

    plt.subplot(121)
    plt.plot(sol_rigid.t/(86400*365.25), sol_rigid.y[0]/r0, 'g-', label='PFDT Rigid')
    plt.plot(sol_gr.t/(86400*365.25), sol_gr.y[0]/r0, 'r--', label='GR c-Lag')
    plt.axhline(1.0, color='k', ls=':')
    plt.xlabel('Years'); plt.ylabel('r / AU')
    plt.title('Orbit Stability: Zero Decay = PFDT Win')
    plt.legend(); plt.grid()

    plt.subplot(122)
    plt.plot(sol_rigid.y[1]*180/np.pi % 360, sol_rigid.y[0]/r0, 'g-', label='Rigid')
    plt.plot(sol_gr.y[1]*180/np.pi % 360, sol_gr.y[0]/r0, 'r--', label='c-Lag')
    plt.xlabel('Angle (deg)'); plt.ylabel('r / AU')
    plt.title('Trajectory: No Spiral-Out')
    plt.legend(); plt.grid()

    plt.tight_layout()
    plt.savefig('rigid_rod_test.png', dpi=150)
    plt.show()

    # Metric: perihelion advance (GR predicts 3.8 arcsec/century)
    delta_r_rigid = np.ptp(sol_rigid.y[0])/r0 * 1e6  # ppm
    delta_r_gr = np.ptp(sol_gr.y[0])/r0 * 1e6

    print(f"PFDT Rigid Delta-r: {delta_r_rigid:.3f} ppm (Your 77-Strap Match?)")
    print(f"GR c-Lag Delta-r:   {delta_r_gr:.3f} ppm")
    print("PFDT WIN" if abs(delta_r_rigid) < abs(delta_r_gr) else "GR HOLDS")

    np.savetxt('rigid_rod_metrics.csv',
               np.column_stack([sol_rigid.t, sol_rigid.y[0], sol_gr.y[0]]),
               header='time,r_rigid,r_gr', delimiter=',')

# EXECUTE CHAIN
rigid_rod_test()
```
