```python
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

# Constants
G, c, h = 6.674e-11, 3.0e8, 6.626e-34
date_stamp = "March 30, 2026"

plt.figure(figsize=(24, 18))
plt.style.use('dark_background')
ax = plt.gca()

# 1. Base Grids & Scales
r_grid = np.logspace(-42, 34, 1000)

# --- 2. DARK MATTER (Galaxy Halo Excess) ---
# Yellow dashed line showing the 5x mass jump at galaxy scales
dm_r = np.logspace(20, 24, 100)
dm_m = 1e42 * (dm_r/1e21)**0.8
plt.loglog(dm_r, dm_m, color='yellow', ls='--', lw=4,
           label='DARK MATTER (Galaxy Halos)', alpha=0.9, zorder=30)

# --- 3. DARK ENERGY (Uniform Density Floor) ---
# Purple line showing the 6e-27 kg/m³ threshold
de_density = 6e-27
de_m = de_density * (4/3 * np.pi * r_grid**3)
plt.loglog(r_grid, de_m, color='purple', lw=4, ls='-',
           label='DARK ENERGY (Uniform)', alpha=0.7, zorder=20)
plt.fill_between(r_grid, 1e-48, de_m, color='purple', alpha=0.1)

# 4. Physics Boundaries
plt.loglog(r_grid, (c**2 * r_grid)/(2*G), color='white', lw=4, ls='--', label='SCHWARZSCHILD', zorder=40)
plt.loglog(r_grid, h/(r_grid*c), color='#ff00ff', lw=4, ls=':', label='COMPTON', zorder=40)

# 5. Objects & Levers
points = {
    "Atoms": (1e-10, 1e-26, "#00ffff"), "Planets": (1e7, 1e25, "#00ffff"),
    "Stars": (1e9, 1e30, "#00ffff"), "Galaxies": (1e21, 1e42, "#00ffff"),
    "Clusters": (1e23, 1e45, "#00ffff"), "MICRO (11)": (1.6e-35, 2.2e-8, "#ffff00"),
    "FLEA (30)": (1e-3, 1e-5, "#ffff00"), "LUNGS": (1, 70, "#ffff00"),
    "MACRO (19)": (4.4e26, 1e54, "#ffff00")
}

for name, (r, m, color) in points.items():
    plt.scatter(r, m, c=color, s=500, edgecolors='white', zorder=100)
    # Staggered offsets for clarity
    y_off = 10**3.5 if "MACRO" in name else 10**1.5
    plt.text(r, m * y_off, name, color=color, fontsize=16, fontweight='bold', ha='center')

# Universe Gold Star
plt.scatter(4.4e26, 1e54, c='gold', marker='*', s=5000, ec='white', lw=3, zorder=110)

# Branding & Labels
plt.title(f"LatticeLogOS: Dark Matter & Energy Unification | {date_stamp}", fontsize=34, color='gold', pad=20)
plt.xlabel("Scale (Meters)", fontsize=22)
plt.ylabel("Mass (Kilograms)", fontsize=22)
plt.xlim(1e-42, 1e34)
plt.ylim(1e-48, 1e110)
plt.grid(True, which="both", ls="-", alpha=0.05)

# Legend Layout
plt.legend(loc='upper left', fontsize=18, frameon=True, facecolor='black', edgecolor='white')

plt.tight_layout()
plt.savefig('latticelog_dark_universe.png', dpi=300, bbox_inches='tight', facecolor='black')
files.download('latticelog_dark_universe.png')
```
